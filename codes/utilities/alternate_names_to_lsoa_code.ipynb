Download the data that match LSOA names to codes from https://www.healthmapswales.wales.nhs.uk/IAS/dataviews/view?viewId=98

```python
import pandas as pd
import numpy as np
```

To get this to work, I downloaded it in Excel format from the above website, loaded in into microsoft excel, and then saved it in xls format. It doesn't work in the form you download it from the website as it says something about a unicode error. I have no idea why this fixes it as that should not have changed anything. 

```python
data = pd.read_excel('code_to_name.xls')
data = data.drop(0)
data = data.drop(columns=['Unnamed: 2','filter','filter.1'])
data = data.rename(columns={'Unnamed: 1' : 'LSOA name','polygon' : 'LSOA code'})
data = data.drop(data.index[1909:])
#data = data.set_index('LSOA code')

data
```

```python
name_to_code = dict(zip(data['LSOA name'], data['LSOA code']))
code_to_name = {name_to_code[code] : code for code in name_to_code.keys()}
```

```python
np.save('name_to_code_dict',name_to_code)
np.save('code_to_name_dict',code_to_name)
```
